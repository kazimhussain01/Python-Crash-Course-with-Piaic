```python
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int)-> int:
    return s1 + s2 + s3 + s4 + s5

df1['Obtained'] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), axis=1)
df1
```

```python
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int)-> tuple[int, int, float, str]:
    Total : int = 500
    Obtained : int = s1 + s2 + s3 + s4 + s5
    per : float = Obtained / Total * 100
    grade : str = ""
    if per >= 80:
        grade = "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    return Total, Obtained, per, grade

df1[['Total', 'Obtained', 'Percentage', 'Grade']] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), axis=1, result_type="expand")
df1
```

## **Map Function for Remark**

```python
from typing import Dict

remark : Dict[str, str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F" : "Fail"
}

df1['Remark'] = df1.Grade.map(remark)
df1
```
